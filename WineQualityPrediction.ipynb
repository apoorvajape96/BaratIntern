```python
df['quality'].unique()
```

```python
df['quality']=[1 if x>=7 else 0 for x in df['quality']]
```

Data Preprocessing


```python
#SPLITTING FEATURES AND TARGET
X = df.drop('quality',axis=1)
```

```python
print(X)
```

```python
y = df['quality']
```

```python
print(y)
```

**Handling** **Imbalanced** **Data**

```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

SPLITTING DATA INTO TRANING AND TESTING DATA

```python
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.2,random_state=2)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

MODEL TRANING



LINEAR REGRASSION MODEL

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

MODEL EVALUTION

```python
#pridiction on training data
training_data_prediction = regressor.predict(X_train)
```

```python
# R square value
from sklearn import metrics
r2_train = metrics.r2_score(y_train,training_data_prediction)
print('R square value:',r2_train)
```

```python
#pridiction on training data
test_data_prediction = regressor.predict(X_test)
```

```python
# R square value
r2_test = metrics.r2_score(y_test,test_data_prediction)
print('R square value:',r2_test)
```

BUILDING PREDICTIVE MODEL

```python
input_data = (   8.5,	0.28,	0.56,	1.8,	0.092,	35,	103,	0.9969,	3.3,	0.75,	10.5)

input_data_as_numpy_array = np.asarray(input_data)

input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = regressor.predict(input_data_reshaped)
print(prediction)
```
